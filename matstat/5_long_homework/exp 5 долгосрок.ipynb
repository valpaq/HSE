```python
import numpy as np
from scipy import stats
```

```python
L=0.5
q5=[]
q10=[]
q100=[]
q1000=[]
q100000=[]
for n in [5,10,100,1000,100000]:
    for j in range(5):
        X=[]
        for i in range(n):
            X.append(-1/L * np.log(np.random.uniform())) 
        if n<100:
            print(X)
        if n==5:
            q5.append(X)
        elif n==10:
            q10.append(X)
        elif n==100:
            q100.append(X)
        elif n==1000:
            q1000.append(X)
        elif n==100000:
            q100000.append(X)
```

```python
q55=np.asarray(q5)
q510=np.asarray(q10)
q5100=np.asarray(q100)
q51000=np.asarray(q1000)
q5100000=np.asarray(q100000)
```

```python
L=4
q5=[]
q10=[]
q100=[]
q1000=[]
q100000=[]
for n in [5,10,100,1000,100000]:
    for j in range(5):
        X=[]
        for i in range(n):
            X.append(-1/L * np.log(np.random.uniform())) 
        if n<100:
            print(X)
        if n==5:
            q5.append(X)
        elif n==10:
            q10.append(X)
        elif n==100:
            q100.append(X)
        elif n==1000:
            q1000.append(X)
        elif n==100000:
            q100000.append(X)
```

```python
q25=np.asarray(q5)
q210=np.asarray(q10)
q2100=np.asarray(q100)
q21000=np.asarray(q1000)
q2100000=np.asarray(q100000)
```

```python
from scipy import stats
import numpy as np
print('Выборки, для которых гипотеза должна быть принята')
print('Уровень значимости 0.01')
t0=0.5
t1=4
for query,size in zip([q510,q5100,q51000,q5100000],[10,100,1000,100000]):
    print('Объем выборки',size)
    crit1 = stats.chi2.ppf(q=0.99, df=size*2)
    for vub in query:
        c = crit1 * (t1-t0) - 2*t0*(size*(np.log(t1) - np.log(t0)))
        chiselka = (size * np.log(t0/t1) ) + (t1 - t0) * np.sum(vub)
        print('Значение с {:.4f}, значение правдоподобия {:.6f}'.format(c, chiselka))
        #Все по формулам, но c не работает, 0 работает, есть подозрения про определенные лямбда, но не уверен полностью
        if 0 < chiselka:
            print('H0 принимается' )
        else:
            print('H0 отклоняется')
```
