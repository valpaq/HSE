```python
import numpy as np
from scipy import stats
```

```python
L=0.5
q5=[]
q10=[]
q100=[]
q1000=[]
q100000=[]
for n in [5,10,100,1000,100000]:
    for j in range(5):
        X=[]
        for i in range(n):
            X.append(-1/L * np.log(np.random.uniform())) 
        if n<100:
            print(X)
        if n==5:
            q5.append(X)
        elif n==10:
            q10.append(X)
        elif n==100:
            q100.append(X)
        elif n==1000:
            q1000.append(X)
        elif n==100000:
            q100000.append(X)
```

```python
q55=np.asarray(q5)
q510=np.asarray(q10)
q5100=np.asarray(q100)
q51000=np.asarray(q1000)
q5100000=np.asarray(q100000)
```

```python
L=1.5
q5=[]
q10=[]
q100=[]
q1000=[]
q100000=[]
for n in [5,10,100,1000,100000]:
    for j in range(5):
        X=[]
        for i in range(n):
            X.append(-1/L * np.log(np.random.uniform())) 
        if n<100:
            print(X)
        if n==5:
            q5.append(X)
        elif n==10:
            q10.append(X)
        elif n==100:
            q100.append(X)
        elif n==1000:
            q1000.append(X)
        elif n==100000:
            q100000.append(X)
```

```python
q155=np.asarray(q5)
q1510=np.asarray(q10)
q15100=np.asarray(q100)
q151000=np.asarray(q1000)
q15100000=np.asarray(q100000)
```

```python
from scipy import stats
import numpy as np
print('Выборки, для которых гипотеза должна быть принята')
print('Уровень значимости 0.05')
t0=0.5
t1=1.5
for query,size in zip([q510,q5100,q51000,q5100000],[10,100,1000,100000]):
    print('Объем выборки',size)
    for vub in query:
        crit = stats.gamma.ppf(0.95, a = len(vub), loc=0, scale=1/np.mean(vub))
        c = -crit * (t0 - t1) + (size * (np.log(t0) - np.log(t1)))
        chiselka = (size * np.log(t0 / t1) ) - (t0 - t1) * np.sum(vub)
        print('Значение ln(с) {:.4f}, значение ln правдоподобия {:.6f}'.format(c, chiselka))
        if c < chiselka:
            print('H0 принимается' )
        else:
            print('H0 отклоняется')
```
